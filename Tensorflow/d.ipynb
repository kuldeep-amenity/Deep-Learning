```python
import math

def sigmoid(x):
    return 1/(1+math.exp(-x))


```

```python
import math

def tanh(x):
    return (math.exp(x) - math.exp(-x)) / (math.exp(x) + math.exp(-x))
```

```python
print(tanh(-2))
print(tanh(0))
print(tanh(2))
```

```python
def relu(x):
    return max(0, x)

# Example
print(relu(-2))  # Output: 0
print(relu(0))   # Output: 0
print(relu(3))   # Output: 3
```
