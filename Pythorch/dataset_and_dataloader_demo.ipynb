```python
from sklearn.datasets import make_classification # Import the make_classification function from sklearn.datasets to create a synthetic dataset.
import torch # Import the PyTorch library, which is essential for building and training neural networks.
```

```python
# Step 1: Create a synthetic classification dataset using sklearn
X, y = make_classification(
    n_samples=10,       # Number of samples (data points) to generate
    n_features=2,       # Number of features (input variables) for each sample
    n_informative=2,    # Number of informative features, which are useful for classification
    n_redundant=0,      # Number of redundant features, which are linear combinations of informative features
    n_classes=2,        # Number of target classes (e.g., binary classification with 0 and 1)
    random_state=42     # Seed for reproducibility, ensuring the same dataset is generated each time
)
```

```python
X # Display the features (input data) of the generated dataset.
```

```python
X.shape # Display the shape (dimensions) of the features array (number of samples, number of features).
```

```python
y # Display the target labels (output data) of the generated dataset.
```

```python
y.shape # Display the shape (dimensions) of the labels array (number of samples).
```

```python
# Convert the data to PyTorch tensors
X = torch.tensor(X, dtype=torch.float32) # Convert the NumPy array X to a PyTorch tensor with float32 data type.
y = torch.tensor(y, dtype=torch.long) # Convert the NumPy array y to a PyTorch tensor with long (integer) data type, suitable for class labels.
```

```python
X # Display the PyTorch tensor X, which now holds the features.
```

```python
y # Display the PyTorch tensor y, which now holds the labels.
```

```python
from torch.utils.data import Dataset, DataLoader # Import Dataset and DataLoader from torch.utils.data to create custom datasets and handle data loading in batches.
```

```python
class CustomDataset(Dataset):
  # Define a custom dataset class that inherits from torch.utils.data.Dataset.
  # This class provides an interface for accessing data samples.

  def __init__(self, features, labels):
    # Constructor method to initialize the dataset with features and labels.
    self.features = features # Store the input features.
    self.labels = labels     # Store the corresponding labels.

  def __len__(self):
    # Returns the total number of samples in the dataset.
    return self.features.shape[0] # The number of samples is the first dimension of the features tensor.

  def __getitem__(self, index):
    # Retrieves a single sample and its label at the given index.
    return self.features[index], self.labels[index] # Return the feature and label for the specified index.
```

```python
dataset = CustomDataset(X, y) # Create an instance of the CustomDataset using the PyTorch tensors X (features) and y (labels).
```

```python
len(dataset) # Get the total number of samples in the dataset using the __len__ method.
```

```python
dataset[2] # Access the third sample (index 2) from the dataset, which calls the __getitem__ method.
```

```python
dataloader = DataLoader(dataset, batch_size=2, shuffle=False) # Create a DataLoader instance from the custom dataset.
# batch_size=2: Specifies that data will be loaded in batches of 2 samples.
# shuffle=False: Data will not be shuffled, so samples are loaded in their original order.
```

```python
for batch_features, batch_labels in dataloader:
  # Iterate through the DataLoader, which yields batches of features and labels.
  print(batch_features) # Print the features tensor for the current batch.
  print(batch_labels)   # Print the labels tensor for the current batch.
  print("-"*50)       # Print a separator line for better readability between batches.
```
