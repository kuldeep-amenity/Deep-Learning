```python
import tensorflow as tf

# --- Configuration ---
# 1. Set the path to your saved .keras model file.
KERAS_MODEL_PATH = 'best_frozen__EfficientNetB3Hybrid_FreezUnFreezModel.keras'

# 2. Set the path where you want to save the converted .tflite model.
TFLITE_MODEL_PATH = 'model.tflite'

print(f"Loading Keras model from: {KERAS_MODEL_PATH}...")

# Load the trained Keras model
try:
    model = tf.keras.models.load_model(KERAS_MODEL_PATH)
except Exception as e:
    print(f"Error loading model: {e}")
    exit()

print("Model loaded successfully.")

# Initialize the TFLiteConverter with the Keras model.
converter = tf.lite.TFLiteConverter.from_keras_model(model)

print("Converting model to TensorFlow Lite format...")

# Convert the model.
tflite_model = converter.convert()

# Save the converted model to the specified file.
with open(TFLITE_MODEL_PATH, 'wb') as f:
    f.write(tflite_model)

print(f" Model successfully converted and saved to: {TFLITE_MODEL_PATH}")
```
